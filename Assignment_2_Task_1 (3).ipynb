In this activity you will practice working with NumPy arrays using a dataset of product sales across two stores over four weeks.

**Instructions:** Fill in the missing code wherever you see `# Your code here`.

```python
import numpy as np

# Store A weekly sales: 4 weeks x 4 product categories
# Columns: Electronics, Clothing, Food, Toys
store_a = np.array([[320, 150, 480, 210],
                    [290, 175, 510, 195],
                    [410, 130, 460, 250],
                    [375, 160, 495, 230]])

# Store B weekly sales: 4 weeks x 4 product categories
store_b = np.array([[280, 200, 430, 190],
                    [310, 185, 450, 220],
                    [390, 170, 475, 205],
                    [350, 195, 440, 240]])

print("Store A Weekly Sales (rows=weeks, cols=Electronics/Clothing/Food/Toys):")
print(store_a)
print("\nStore B Weekly Sales (rows=weeks, cols=Electronics/Clothing/Food/Toys):")
print(store_b)
```

---
## Part 1: Statistical Analysis

Using `store_a`, complete the following tasks:

### Task 1.1: Calculate the overall mean of all sales in Store A

```python

mean_all = np.mean(store_a)

print("Overall Mean Sales:", mean_all)
```

### Task 1.2: Find the minimum and maximum sales values in Store A

```python

min_sales =np.min(store_a)
max_sales =np.max(store_a)

print("Minimum Sales:", min_sales)
print("Maximum Sales:", max_sales)
```

### Task 1.3: Calculate the total sales per product category across all 4 weeks

*Hint: Use the `axis` parameter. axis=0 operates down columns (across rows), axis=1 operates across columns (across a single row).*

```python

total_per_category = np.sum(store_a, axis=0)

print("Total sales per category (Electronics, Clothing, Food, Toys):", total_per_category)
```

### Task 1.4: Calculate the total sales per week (sum across all 4 product categories for each week)

```python

total_per_week = np.sum(store_a, axis=1)

print("Total sales per week:", total_per_week)
```

### Task 1.5: Calculate the variance and standard deviation of Store A sales

```python

variance = np.var(store_a)
std_dev = np.std(store_a)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

## Part 2: Array Manipulation

Using `store_a`, complete the following tasks:

### Task 2.1: Reshape the array to have 8 rows and 2 columns

```python

reshaped_8x2 = store_a.reshape(8,2)

print("Reshaped to 8x2:")
print(reshaped_8x2)
```

### Task 2.2: Find the total number of elements in the array

```python

total_elements = store_a.size

print("Total elements:", total_elements)
```

### Task 2.3: Flatten the array into a 1D array

```python

flattened = store_a.flatten()

print("Flattened array:", flattened)
```

### Task 2.4: Transpose the array (swap rows and columns)

```python

transposed = store_a.T

print("Transposed array:")
print(transposed)
```

### Task 2.5: Sort the flattened array in ascending order

*Hint: Use `np.sort()`*

```python

sorted_array = np.sort(flattened)

print("Sorted sales values:", sorted_array)
```

## Part 3: Operations Between Two Arrays

Using both `store_a` and `store_b`, complete the following:

### Task 3.1: Add the two arrays element-wise to get combined sales

```python

combined_sales = store_a + store_b

print("Combined Sales (Store A + Store B):")
print(combined_sales)
```

### Task 3.2: Subtract Store B from Store A to see the sales difference per cell

```python

sales_diff = store_a - store_b

print("Sales Difference (Store A - Store B):")
print(sales_diff)
```

### Task 3.3: Use boolean indexing to find all cells in `store_a` where sales exceeded 400

```python

high_sales = store_a[store_a > 400]

print("Sales values above 400:", high_sales)
```

### Task 3.4: Perform matrix multiplication between `store_a` and the transpose of `store_b`

*Hint: Use `np.dot()` and `.T` for transpose*

```python

matrix_mult = np.dot(store_a, store_b.T)

print("Matrix Multiplication (A x B^T):")
print(matrix_mult)
```
